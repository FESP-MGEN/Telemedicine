```python
# Import packages
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Merge two dataframe: NLTK and TextBlob indicators
NLTK_TextBlob = pd.merge(NLTK_base, TextBlob_base, on='ID_Abstract')

# Import the dataframe with the information "dominante_intervention"
df_excel = pd.read_excel (r'C:\\Users\\....\\Database\\Dominance_of_telemedicine.xlsx')

# Change column names
df_excel.columns = ['ID_2', 'Dominant intervention', 'ID_Abstract'] 

# Change the format of NLTK_TextBlob.ID_Abstract into int64
NLTK_TextBlob.ID_Abstract = NLTK_TextBlob.ID_Abstract.astype(np.int64)

# Merge two dataframe: sentiment analysis result and the excel file with "dominante_information"
NLTK_TextBlob_domi = pd.merge(NLTK_TextBlob, df_excel, on='ID_Abstract')
```

```python
# Create two databases for graphics
domi_yes = NLTK_TextBlob_domi.loc[NLTK_TextBlob_domi['Dominant intervention'] == "Yes"]
domi_no = NLTK_TextBlob_domi.loc[NLTK_TextBlob_domi['Dominant intervention'] == "No"]
```

```python
###########################################################################################
# Scatter plot
###########################################################################################
```

```python
# Graphic of Polarity and Compound according to the Dominance of the intervention
##################################################################################
G1=plt.subplots(figsize=(20, 15))


G1=plt.subplot(2, 2, 1)
# Subjectivity Polarity 
x1 = sns.regplot(x="Polarity", y="Subjectivity", data=domi_yes,  color='#850606')
plt.legend(['Dominant intervention = Yes'],loc=1,prop={'size':10})
plt.xlabel('Polarity', size = 11)
plt.ylabel('Subjectivity', size = 11)
plt.ylim(-0.01, 1)


G1=plt.subplot(2, 2, 2)
# Subjectivity Polarity 
x2 = sns.regplot(x="Polarity", y="Subjectivity", data=domi_no,  color='#6EAED2')
plt.legend(['Dominant intervention = No'],loc=1,prop={'size':10})
plt.xlabel('Polarity', size = 11)
plt.ylabel('', size = 11)
plt.ylim(-0.01, 1)


G1=plt.subplot(2, 2, 3)
# Subjectivity Compound 
x1 = sns.regplot(x="Compound", y="Subjectivity", data=domi_yes,  color='#850606')
plt.legend(['Dominant intervention = Yes'],loc=1,prop={'size':10})
plt.xlabel('Compound', size = 11)
plt.ylabel('Subjectivity', size = 11)
plt.ylim(-0.01, 1)


G1=plt.subplot(2, 2, 4)
# Subjectivity Compound 
x2 = sns.regplot(x="Compound", y="Subjectivity", data=domi_no,  color='#6EAED2')
plt.legend(['Dominant intervention = No'],loc=1,prop={'size':10})
plt.xlabel('Compound', size = 11)
plt.ylabel('', size = 11)
plt.ylim(-0.01, 1)


plt.tight_layout()

# Print the plot
plt.show() 

# Save the plot
fig = G1.get_figure()
fig.savefig("C:\\Users\\....\\Result\\Graph_Scatter_plot_1.png")
```

```python
# Graphic of Positivity, Negativity and Neutrality according to the dominance of the intervention
################################################################################################

G2=plt.subplots(figsize=(20, 20))


G2=plt.subplot(3, 2, 1)
# Subjectivity Positivity 
x1 = sns.regplot(x="Positive", y="Subjectivity", data=domi_yes, color='#850606')
plt.legend(['Dominant intervention = Yes'],loc=1,prop={'size':10})
plt.xlabel('Positivity', size = 11)
plt.ylabel('Subjectivity', size = 11)
plt.ylim(-0.01, 1)


G2=plt.subplot(3, 2, 2)
# Subjectivity Positivity 
x2 = sns.regplot(x="Positive", y="Subjectivity", data=domi_no, color='#6EAED2')
plt.legend(['Dominant intervention = No'],loc=1,prop={'size':10})
plt.xlabel('Positivity', size = 11)
plt.ylabel('', size = 11)
plt.ylim(-0.01, 1)


G2=plt.subplot(3, 2, 3)
# Subjectivity Neutral
x1 = sns.regplot(x="Neutral", y="Subjectivity", data=domi_yes, color='#850606')
plt.legend(['Dominant intervention = Yes'],loc=1,prop={'size':10})
plt.xlabel('Neutrality', size = 11)
plt.ylabel('Subjectivity', size = 11)
plt.ylim(-0.01, 1)


G2=plt.subplot(3, 2, 4)
# Subjectivity Neutral
x2 = sns.regplot(x="Neutral", y="Subjectivity", data=domi_no, color='#6EAED2')
plt.legend(['Dominant intervention = No'],loc=1,prop={'size':10})
plt.xlabel('Neutrality', size = 11)
plt.ylabel('', size = 11)
plt.ylim(-0.01, 1)


G2=plt.subplot(3, 2, 5)
# Subjectivity Negativity
x1 = sns.regplot(x="Negative", y="Subjectivity", data=domi_yes, color='#850606')
plt.legend(['Dominant intervention = Yes'],loc=1,prop={'size':10})
plt.xlabel('Negativity', size = 11)
plt.ylabel('Subjectivity', size = 11)
plt.ylim(-0.01, 1)


G2=plt.subplot(3, 2, 6)
# Subjectivity Negativity
x2 = sns.regplot(x="Negative", y="Subjectivity", data=domi_no, color='#6EAED2')
plt.legend(['Dominant intervention = No'],loc=1,prop={'size':10})
plt.xlabel('Negativity', size = 11)
plt.ylabel('', size = 11)
plt.ylim(-0.01, 1)


plt.tight_layout()

# Print the plot
plt.show() 

# Save the plot
fig = G2.get_figure()
fig.savefig("C:\\Users\\....\\Result\\Graph_Scatter_plot_2.png")
```

```python
G3=plt.subplots(figsize=(30, 30))


G3=plt.subplot(3, 2, 1)
# Subjectivity
sns.kdeplot(domi_yes['Subjectivity'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Subjectivity'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=1, prop={'size':17})
plt.xlabel('Subjectivity', size = 18)
plt.ylabel('Density', size = 18)


G3=plt.subplot(3, 2, 2)
# Polarity
sns.kdeplot(domi_yes['Polarity'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Polarity'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=1, prop={'size':17})
plt.xlabel('Polarity', size = 18)


G3=plt.subplot(3, 2, 3)
# Compound
sns.kdeplot(domi_yes['Compound'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Compound'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=2, prop={'size':17})
plt.xlabel('Compound', size = 18)
plt.ylabel('Density', size = 18)


G3=plt.subplot(3, 2, 4)
# Neutrality
sns.kdeplot(domi_yes['Neutral'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Neutral'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=2, prop={'size':17})
plt.xlabel('Neutrality', size = 18)


G3=plt.subplot(3, 2, 5)
# Positivity
sns.kdeplot(domi_yes['Positive'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Positive'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=1, prop={'size':17})
plt.xlabel('Positivity', size = 18)
plt.ylabel('Density', size = 18)


G3=plt.subplot(3, 2, 6)
# Negativity
sns.kdeplot(domi_yes['Negative'], label="Dominant intervention = Yes", color="#850606", linewidth=3)
sns.kdeplot(domi_no['Negative'], label="Dominant intervention = No", color="#6EAED2", linewidth=3)
plt.legend(loc=1, prop={'size':17})
plt.xlabel('Negativily', size = 18)

plt.tight_layout()

# Print the plot
plt.show() 

# Save the plot
fig = G3.get_figure()
fig.savefig("C:\\Users\\....\\Result\\Graph_Kernel_curve.png")
```
