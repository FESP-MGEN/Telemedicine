```python
# Import packages (Download and install matplotlib_venn_wordcloud-0.2.4-py2.py3-none-any.whl)
from matplotlib_venn import venn3, venn3_circles #venn diagram
import matplotlib as plt #venn diagram

# Import packages (Download and install plotly-4.5.4-py2.py3-none-any.whl)
import plotly.graph_objects as go #bubble plot
import plotly #bubble plot
```

```python
################################################################################################
# Venn graphic
################################################################################################

v = venn3(subsets = (1702,1155,44,354,28,128,138), set_labels = ('Pubmed','Cochrane','Embase'), alpha = 0.5)

# Coulours of the circles
v.get_patch_by_id('100').set_color('#BBDAEA') 
v.get_patch_by_id('010').set_color('#C94741')
v.get_patch_by_id('001').set_color('#FACAB1') 
v.get_patch_by_id('011').set_color('#F19E7D') 
v.get_patch_by_id('101').set_color('#6EAED2') 
v.get_patch_by_id('110').set_color('#850606') 
v.get_patch_by_id('111').set_color('#0A3B70') 

# Adjust labels from counts to names
v.get_label_by_id('100').set_text('1702 \n (d=2)') 
v.get_label_by_id('010').set_text('1155 \n (d=82)')
v.get_label_by_id('001').set_text('354 \n (d=13)')
v.get_label_by_id('011').set_text('128')
v.get_label_by_id('101').set_text('28')
v.get_label_by_id('110').set_text('44')
v.get_label_by_id('111').set_text('138') 

# Circles
c = venn3_circles(subsets = (1702,1155,44,354,28,128,138), linewidth=1, color='white')
c[0].set_lw(1.0)
c[1].set_lw(1.0)
c[2].set_lw(1.0)

# Save the plot
plt.pyplot.savefig("C:\\Users\\....\\Result\\Graph_Venn_diagram.png")
```

```python
################################################################################################
# Bubble plot
################################################################################################

size = [1, 0, 1, 43, 24, 7, 39, 6, 36]

fig = go.Figure(data=[go.Scatter(
    x=[-1, -1, -1, 0, 0, 0, 1, 1, 1],
    y=[-1, 0, 1, -1, 0, 1, -1, 0, 1],
    mode='markers+text',
    text=["0.6", "0", "0.6", "27.4", "15.3", "4.5", "24.8", "3.8", "22.9"],
    textposition="middle right", 
    textfont=dict(
        family="Arial, sans-serif",
        size=14,
        color="Black"
    ),
    
        marker=dict(
        size=size,
        sizemode='area',
        sizeref=1.*max(size)/(40.**2),
        sizemin=4,
        opacity=[0.8, 0.8, 0.8, 0.8, 0.8, 0.8, 0.8, 0.8, 0.8],
        color=['#FACAB1', '#850606', '#850606',
               '#6EAED2', '#FACAB1', '#850606', 
               '#6EAED2', '#6EAED2', '#FACAB1']
    )   
 
)])

fig.update_layout(
    annotations=[
        go.layout.Annotation(
            x=0.5,
            y=-0.15,
            showarrow=False,
            text="Consequences",
            height=20,   
            xref="paper",
            yref="paper",
            font=dict(
                family="Arial, sans-serif",
                size=17,
                color="black",
            ),
        ),
        go.layout.Annotation(
            x=-0.07,
            y=0.5,
            showarrow=False,
            text="Cost",
            height=20,   
            textangle=-90,
            xref="paper",
            yref="paper",
            font=dict(
                family="Arial, sans-serif",
                size=17,
                color="black",
            ),
        )        
    ],
    autosize=True,
    margin=dict(
        b=100
    ),
    
    height=750, 
    width=790, 
)
           
# Options for x-axis and y-axis
fig.update_xaxes(showline=True, linewidth=1, linecolor='black', showgrid=False, zeroline=False,
                 tickvals = [-1,0,1], 
                 ticktext =["Negative", "Neutral", "Positive"],
                 tickfont = dict(
                     family = 'Arial, sans-serif', 
                     size = 15, 
                     color = 'black')
                )

fig.update_yaxes(showline=True, linewidth=1, linecolor='black', showgrid=False, zeroline=False, 
                 tickvals = [-1,0,1], 
                 ticktext =["Negative", "Neutral", "Positive"],
                 tickangle = -90,
                 tickwidth=2,
                 tickfont = dict(
                     family = 'Arial, sans-serif', 
                     size = 15, 
                     color = 'black')
                )

fig.update_layout(plot_bgcolor='white') 

# Print the plot
fig.show()

# Save the plot
plotly.offline.plot(fig, filename='C:\\Users\\....\\Result\\Graph_Bubble_plot')
```
